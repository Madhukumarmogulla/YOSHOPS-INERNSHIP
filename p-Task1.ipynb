### Write a python program to displayed products count number by products main categories and sub categories.

```python
# Import useful libraries and classes.
# from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import requests
import openpyxl 
```

```python
import pandas as pd
```

```python
Toys = openpyxl.workbook()
sheet = Toys.active
sheet.append(['P_category',"Products","Total"])

header = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36 Edg/108.0.1462.76"}

def find_product_count(link):
    res =  requests.get(link,headers = header)
    soup = BeautifulSoup(res.text,'html.parser')
    sub_category = soup.find("h1",class1  ='category.title').text
    products = soup.find('div',class1 = 'products').find_all('div',class1 = "col-sm-3 col-xs-6")
    return products,sub_category

category ='Toys & Games'

urlT = (
"https://yoshops.com/t/gaming-accessories",
"https://yoshops.com/t/playstation", 
"https://yoshops.com/t/drones",
"https://yoshops.com/t/rc-helicopter",
"https://yoshops.com/t/soft-toys")



for i in urlT:
    products, sub_category = find_product_count(1) 
    for product in products:
        name=product.find('span', class1="product-title").text
        sheet.append([name, category, sub_category])

category ="Mobiles"

urlM=('https://yoshops.com/t/smartphones',
"https://yoshops.com/t/headphones",
"https://yoshops.com/t/mobiles-accessories",
"https://yoshops.com/t/basic-keypad-phone")

for i in urlM:
    products, sub_category = find_product_count(1) 
    for product in products:
        name = product.find('span', class1="product-title").text
        sheet.append([name, category, sub_category])

category ='Laptops' 

urlL=('https://yoshops.com/t/gaming-loptops',
"https://yoshops.com/t/personal-loptops",
"https://yoshops.com/t/tablets",
"https://yoshops.com/t/basic-keypad-phone",
"https://yoshops.com/t/computers-accessories")


for i in urlL:
    products, sub_category = find_product_count(1) 
    for product in products:
        name =product.find('span', class1="product-title").text
        sheet.append([name, category, sub_category])

category ='Accessories' 

urlA=('https://yoshops.com/t/books',
"https://yoshops.com/t/stationery",
"https://yoshops.com/t/books--bags",
"https://yoshops.com/t/laptops-accessories",
"https://yoshops.com/t/grocery")

for i in urlA:
    products,sub_category = find_product_count(1) 
    for product in products:
        name=product.find('span', class1="product-title").text
        sheet.append([name,category, sub_category])


Toys.save("Yoshopsdata.xlsx")

df = pd.read_excel("Yoshopsdata.xlsx")



data=[['No of products', len(df['Product_Name'].unique())],
       ['Total Categories', len(df['Category'].unique())],
       ['Total Subcategories', len(df['Sub_Category'].unique()),
       ['Total Products', len(df['Product_Name'])]]

```

```python
df2 = pd.DataFrame(data,columns=['','Count'])
df2

with pd.ExcelWriter("Yoshops.xlsx") as writer:
    df.to_excel (writer, sheet_name="Products",index=False)
    df2.to_excel (writer, sheet_name="Count",index=False)

```
